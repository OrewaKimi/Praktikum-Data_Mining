```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('Data_Mining.csv')

# Tugas 1: Eksplorasi Data dengan Pandas

# Untuk menampilkan 5 data pertama dan terakhir
print("Tampilkan 5 data pertama:")
print(df.head())
print("\nTampilkan 5 data terakhir:")
print(df.tail())

# Menghitung jumlah produk yang terjual
total_penjualan = df['Global_Sales'].sum()
print(f"\nTotal penjualan: {total_penjualan}")

# Mmencari produk dengan harga tertinggi dan terendah
print(f"\nProduk dengan harga tertinggi: {df['Name'][df['Global_Sales'].idxmax()]}")
print(f"Produk dengan harga terendah: {df['Name'][df['Global_Sales'].idxmin()]}")

# Kelompokkan data berdasarkan jenis produk dan hitung total penjualan masing-masing
penjualan_per_genre = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
print("\nTotal penjualan per genre:")
print(penjualan_per_genre)

# Visualisasi: Bar chart penjualan per produk
df.sort_values('Global_Sales', ascending=False).head(10).plot(kind='bar', x='Name', y='Global_Sales', figsize=(12, 6))
plt.title('10 Produk dengan Penjualan Tertinggi')
plt.xlabel('Nama Produk')
plt.ylabel('Total Penjualan')
plt.show()

# Visualisasi: Pie chart proporsi penjualan per genre
plt.figure(figsize=(8, 6))
penjualan_per_genre.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporsi Penjualan per Genre')
plt.axis('equal')
plt.show()

# Tugas 2: Analisis Data dengan Statistik Deskriptif

# Hitung rata-rata, median, dan modus harga produk
print(f"Rata-rata harga: {df['Global_Sales'].mean()}")
print(f"Median harga: {df['Global_Sales'].median()}")
print(f"Modus harga: {df['Global_Sales'].mode()[0]}")

# Hitung standar deviasi harga produk
print(f"Standar deviasi harga: {df['Global_Sales'].std()}")

# Membuat box plot harga produk
df['Global_Sales'].plot(kind='box', figsize=(8, 6))
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga Produk')
plt.show()

# Tugas 3: Membersihkan Data

# Cek missing values
print("Missing values:")
print(df.isna().sum())

# Hapus data dengan missing values
df = df.dropna()

# Ganti missing values dengan nilai rata-rata
df['Global_Sales'] = df['Global_Sales'].fillna(df['Global_Sales'].mean())

# Cek duplikat data
print("\nDuplikat data:")
print(df.duplicated().sum())

# Hapus duplikat data
df = df.drop_duplicates()
```
